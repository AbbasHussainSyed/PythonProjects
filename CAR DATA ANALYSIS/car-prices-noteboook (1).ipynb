```python
plt.xticks(rotation=90)
sns.barplot(x='Category',y='Price',data=train)
```

The above plot shows that **price of Goods Wagon is higher** than every category as Goods Wagon require tough material and good built quality for transportation. 

```python
plt.figure(figsize=(8,5))
plt.xticks(rotation=90)
sns.countplot(x='Category',hue='Leather interior',data=train)
plt.show()
```

From above countplot we can conclude that **most of Sedans and Jeeps have Leather interior** in them.But we can also see that categories like **Microbus,Goods Wagon doesn't have Leather interiors**.

```python
plt.figure(figsize=(8,5))
plt.xticks(rotation=90)
sns.countplot(x='Category',hue='Drive wheels',data=train)
plt.show()
```

From above countplot we can conclude that most of **Jeeps have 4x4 Drive Wheels** as these 4x4 wheels are excellent for offroad drive while mostly **Sedans have Front Drive Wheels** as sedans are used mostly for daily routine usage or for luxorious drive.

```python
plt.figure(figsize=(8,5))
plt.xticks(rotation=90)
sns.countplot(x='Category',hue="Gear box type",data=train)
plt.legend(loc="upper right")
plt.show()
```

The above plot shows us that different categories demands different gear shift box as per usage.We can clearly see that **Jeep,Hatchback,Sedan Categories Uses Automatic the Most** whereas **Microbus,Goods Wagon Categories Used Manual Shifting the most** .

```python
plt.figure(figsize=(8,5))
plt.xticks(rotation=90)
sns.countplot(x='Category',hue="Fuel type",data=train)
plt.legend(loc="upper right")
```

The above plot shows us that most of these categories used **Petrol or Disesel as Fuel Type** We can clearly see that  **Microbus,Goods Wagon Categories Uses Diesel most** showing that other Fuel Types like **Plug-in-Hybrid,LPG,Hydrogen as useless fuel types**.
